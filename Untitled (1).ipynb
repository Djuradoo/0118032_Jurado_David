```python
    #5
#a) cuántas letras vocales en mayúscula se entraron; b) cuántas letras con 
#tilde se entraron (minúsculas y mayúsculas); c) cuántos dígitos se entraron,
#d) cuántos espacios se entraron; e) cuántas palabras reservadas se entraron.
    
Vocales= 'AEIOU'
Tildes = 'ÁÉÍÓÚáéíóú'
Digitos = '0123456789'
Espacios = ' '
Palabras_reservadas = ['and','del','for','is','raise','assert','if','else',\
'elif','from','lambda','return','break','global','not','try','class','except',\
'or','while','continue','exec','import','yield','def','finally','in','print']

#Ahora nombramos los contadores

cont_voc = 0
cont_til = 0
cont_dig = 0
cont_esp = 0
cont_res = 0

#Ingresamos la cadena de texto
cadena = input('Ingrese una cadena de texto: ')

#Vamos a devolver la cadena de texto en una lista que ingresamos a una lista 
#para así poder calcular a partir de los contadores
for i in range (len(cadena)):
    if cadena[i] in Vocales:
        cont_voc += 1
    elif cadena[i] in Tildes:
        cont_til += 1
    elif cadena[i] in Digitos:
        cont_dig += 1
    elif cadena[i] in Espacios:
        cont_esp += 1
    elif cadena[i] in Palabras_reservadas :
        cont_res += 1
shi = cadena.split() #transformamos a lista

for n, i in enumerate(shi): #Se itera en los elementos de la lista y se compara
    if i in Palabras_reservadas:#con las palabras reservadas. Si cumple se suma
        cont_res+=1        #al contador 
        
print(f'Se entraron {cont_voc} letras vocales en mayúscula')
print(f'Se entraron {cont_til} letras con tilde')
print(f'Se entraron {cont_dig} digitos')
print(f'Se entraron {cont_esp} espacios')
print(f'Se entraron {cont_res} palabras reservadas')
```
